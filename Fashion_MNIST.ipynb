```python
import numpy as np

import torch
from torch import optim
from torch import nn
import torch.nn.functional as F
from torchvision import datasets, transforms
from torch import nn


import helper
import time

import matplotlib.pyplot
%matplotlib inline
%config InlineBackend.figure_format ='retina'

```

```python
#define a transform to normalize the data
transform = transforms.Compose([transforms.ToTensor(),transforms.Normalize((0.5,0.5,0.5),(0.5,0.5,0.5))])

#download and load the training data
trainset= datasets.FashionMNIST('F_MNIST_data', download=True, train=True, transform=transform)
trainloader = torch.utils.data.DataLoader(trainset, batch_size=64, shuffle= True)

#download and load the test data
testset= datasets.FashionMNIST('F_MNIST_data', download=True, train=False, transform=transform)
testloader = torch.utils.data.DataLoader(testset, batch_size=64, shuffle= True)
```

```python
#here we can see one of the images
image, label = next(iter(trainloader))
helper.imshow(image[0,:]);
```

```python
#Building the network

class Network(nn.Module):
    def __init__(self):
        super().__init__()
        self.fc1 = nn.Linear(784,400)
        self.fc2 = nn.Linear(400,200)
        self.fc3 = nn.Linear(200,100)
        self.fc4 = nn.Linear(100,10)
    
    def forward(self, x):
        x= self.fc1(x)
        x= F.relu(x)
        x= self.fc2(x)
        x= F.relu(x)
        x= self.fc3(x)
        x= F.relu(x)
        x= self.fc4(x)
        x= F.softmax(x,dim=1)
        
        return x
model = Network()
model
```

```python
#Forward pass through the network
dataiter=iter(trainloader)
images,labels=dataiter.next()

#resize the images into a 1D vector, new shape is (batch size, color channels, image pixels)
images.resize_(images.shape[0],1,784)

#forward pass through the network
img_idx = 0
ps = model.forward(images[img_idx,:])

img = images[img_idx]
helper.view_classify(img.view(1,28,28),ps)
```

```python
#Defining the hyperparameter
input_size= 784
hidden_size=[100,28]
output_size=10

#alternative way to build a feed-forward network
model=nn.Sequential(nn.Linear(input_size,hidden_size[0]),
                    nn.ReLU(),
                    nn.Linear(hidden_size[0],hidden_size[1]),
                    nn.ReLU(),
                    nn.Linear(hidden_size[1],output_size),
                    nn.Softmax(dim=1))
print(model)

#forward pass through the network
images,labels = next(iter(trainloader))
images.resize_(images.shape[0], 1, 784)
ps = model.forward(images[0,:])
helper.view_classify(images[0].view(1, 28, 28), ps)
```

```python
#3rd way of building the model
from collections import OrderedDict
model = nn.Sequential(OrderedDict([
                      ('fc1', nn.Linear(input_size, hidden_size[0])),
                      ('relu1', nn.ReLU()),
                      ('fc2', nn.Linear(hidden_size[0], hidden_size[1])),
                      ('relu2', nn.ReLU()),
                      ('output', nn.Linear(hidden_size[1], output_size)),
                      ('softmax', nn.Softmax(dim=1))]))

#forward pass through the network
images,labels = next(iter(trainloader))
images.resize_(images.shape[0], 1, 784)
ps = model.forward(images[0,:])
helper.view_classify(images[0].view(1, 28, 28), ps)
```

```python
#defining the loss function and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(model.parameters(),lr=0.03)
```

```python
print('Initial weights - ', model.fc1.weight)

images,labels = next(iter(trainloader))
images.resize_(64,784)

#Clear the gradients
optimizer.zero_grad()

#forward pass, then backward pass then update the weights
output =model.forward(images)
loss=criterion(output,labels)
loss.backward()
print('Gradient -', model.fc1.weight.grad)
optimizer.step()

print('Updated weights - ', model.fc1.weight)
```

```python
epochs = 3
print_every = 40
steps = 0

for e in range (epochs):
    running_loss = 0
    for images, labels in iter (trainloader):
        steps+=1
        #Flatten MNIST images into a 784 long vector
        images.resize_(images.size()[0], 784)
        
        optimizer.zero_grad()
            
        # Forward and backward passes
        output = model.forward(images)
        loss = criterion(output, labels)
        loss.backward()
        optimizer.step()
        
        running_loss += loss.item()
        
        if steps % print_every == 0:
            print("Epoch: {}/{}... ".format(e+1, epochs),
                  "Loss: {:.4f}".format(running_loss/print_every))
            
            running_loss = 0
```

```python
images, labels = next(iter(trainloader))

img = images[0].view(1, 784)
# Turn off gradients to speed up this part
with torch.no_grad():
    logits = model.forward(img)

# Output of the network are logits, need to take softmax for probabilities
ps = F.softmax(logits, dim=1)
helper.view_classify(img.view(1, 28, 28), ps, version='Fashion')
```
